We observe that the null values in Promo2SinceWeek, Promo2SinceYesr and PromoInterval appear only when the promo2 was absent, i.e no promotion was done. Hence we fill them with zeroes.

```python
stores['Promo2SinceWeek'].replace(np.nan, 0, inplace=True)
stores['Promo2SinceYear'].replace(np.nan, 0, inplace=True)
stores['PromoInterval'].replace(np.nan, '0', inplace=True)
stores.head(20)
```

```python
stores['PromoInterval'].unique()
```

```python
#promo = {"PromoInterval": 0: 0, 'Jan,Apr,Jul,Oct': 1, 'Feb,May,Aug,Nov': 2, 'Mar,Jun,Sept,Dec': 3}
stores["PromoInterval"].replace({'0':0,'Jan,Apr,Jul,Oct':1, 'Feb,May,Aug,Nov':2,  'Mar,Jun,Sept,Dec': 3}, inplace=True)
stores.head(20)
```

```python
#Filling Null Values for competition distace
stores[stores['CompetitionDistance'].isna()].Store.unique()
```

```python
#understanding distribution of Ciompetition Distance

sns.distplot(stores["CompetitionDistance"],hist=True)
print(stores["CompetitionDistance"].mean())
print(stores["CompetitionDistance"].mode())
print(stores["CompetitionDistance"].median())
```

```python
stores["CompetitionDistance"] = stores["CompetitionDistance"] .transform(lambda x: x.fillna(x.median()))
```

```python
#stores[stores['Store']==622].CompetitionDistance.unique()
#since there is not much info of competition's open since month and year, filling it with median.
stores['CompetitionOpenSinceMonth']=stores['CompetitionOpenSinceMonth'].fillna(stores['CompetitionOpenSinceMonth'].median())
stores['CompetitionOpenSinceYear']=stores['CompetitionOpenSinceYear'].fillna(stores['CompetitionOpenSinceYear'].median())
stores.head()
```

```python
stores.info()
```

```python
# Data transformations
mappings = {'0': 0, 'a':1, 'b':2, 'c':3, 'd':4}
print(stores['StoreType'].unique())
print(stores['Assortment'].unique())

#stores['StoreType'].replace(mappings,inplace=True)
#stores['Assortment'].replace(mappings, inplace=True)
#train_df['StateHoliday'].replace(mappings, inplace=True)
#test_df['StateHoliday'].replace(mappings, inplace=True)
```

```python
stores['StoreType'].replace(mappings,inplace=True)
stores['Assortment'].replace(mappings, inplace=True)
```

```python
###La1

stores = stores.astype({'CompetitionDistance': 'int', 'CompetitionOpenSinceYear': 'int','Promo2SinceWeek': 'int',
                       'CompetitionOpenSinceMonth':'int','Promo2SinceYear':'int'})
```

```python
print(stores.isnull().all())
```

```python
#Merging the Stores dataset with train and test
train_df = pd.merge(train, stores, on='Store',how="left")
test_df = pd.merge(test, stores, on='Store',how="left")

train_df['StateHoliday'].replace(mappings, inplace=True)
test_df['StateHoliday'].replace(mappings, inplace=True)
train_df.head(20)
```

```python
train_copy = train_df.copy( deep=True)
test_copy = test_df.copy( deep=True)
```

```python
print(test_copy.columns)
print(train_copy.columns)
```

```python
train_copy.drop(['Date','Customers', 'Store','Promo2SinceWeek', 'Promo2SinceYear', 'PromoInterval'], axis=1, inplace=True)
test_copy.drop(['Date','Id', 'Store', 'Promo2SinceWeek', 'Promo2SinceYear', 'PromoInterval'], axis=1, inplace=True)
```

```python
X=train_copy.drop('Sales',axis=1)
y=train_copy['Sales']
```

```python
#train test split
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
##### Predicting important features####

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators = 120, random_state=2021,verbose=2,max_depth=10,n_jobs=-1)
rf.fit(X_train, y_train)
y_pred=rf.predict(X_test)
print(mean_squared_error(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```

```python
ser  = pd.Series(rf.feature_importances_)
ser.plot(kind='bar')
print(X_train.columns)
```

Variables  'Promo2SinceWeek', 'Promo2SinceYear', 'PromoInterval' seems to be least important features , hence dropping them (above).

```python
####Aplying ElasticNet model
from sklearn.linear_model import ElasticNet
parameters = dict(alpha=np.linspace(0,20,25),
                  l1_ratio=np.linspace(0,1,25))
from sklearn.model_selection import RandomizedSearchCV
en = ElasticNet()
from sklearn.model_selection import KFold
kfold = KFold(n_splits=5, random_state=2020, shuffle=True)
cv = RandomizedSearchCV(en, param_distributions=parameters,
                  cv=kfold,scoring='neg_mean_squared_error')

cv.fit(X,y)
# Best Parameters
print(cv.best_params_)

print(cv.best_score_)

en_model= cv.best_estimator_
```

```python
Id=test_df['Id']
y_pred_en = en_model.predict(test_copy)
file1 = pd.DataFrame({'Id':Id,'Sales':y_pred_en})
file1.to_csv("elasticnet1.csv",index=False)
```
